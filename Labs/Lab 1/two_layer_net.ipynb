# CIFAR-10 Data Loading
We will now train the implemented neural network in a real dataset, the widely used CIFAR-10.
You can download this dataset at: https://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz
Then, you can put it in the folder malcom/datasets (or anywhere else you want, but then you will need to change the path in the below code).

```python
from malcom.data_utils import load_CIFAR10

def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000):
    """
    This will allows us to load the CIFAR-10 dataset and perform preprocessing to prepare
    it for the two-layer neural net classifier.  
    """
    # Load the raw CIFAR-10 data
    cifar10_dir = 'malcom/datasets/cifar-10-batches-py'
    
    # Cleaning up variables to prevent loading data multiple times (which may cause memory issue)
    try:
       del X_train, y_train
       del X_test, y_test
       print('Clear previously loaded data.')
    except:
       pass

    X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)
        
    # Subsample the data
    mask = list(range(num_training, num_training + num_validation))
    X_val = X_train[mask]
    y_val = y_train[mask]
    mask = list(range(num_training))
    X_train = X_train[mask]
    y_train = y_train[mask]
    mask = list(range(num_test))
    X_test = X_test[mask]
    y_test = y_test[mask]

    # Normalize the data: subtract the mean image
    mean_image = np.mean(X_train, axis=0)
    X_train -= mean_image
    X_val -= mean_image
    X_test -= mean_image

    # Reshape data to rows
    X_train = X_train.reshape(num_training, -1)
    X_val = X_val.reshape(num_validation, -1)
    X_test = X_test.reshape(num_test, -1)

    return X_train, y_train, X_val, y_val, X_test, y_test


# Invoke the above function to get our data.
X_train, y_train, X_val, y_val, X_test, y_test = get_CIFAR10_data()
print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```

# Network Training
To train our network we will use (as before) SGD. In addition, we will adjust the learning rate with an exponential learning rate schedule as optimization proceeds; after each epoch, we will reduce the learning rate by multiplying it by a decay rate.

```python
input_size = 32 * 32 * 3
hidden_size = 50
num_classes = 10
net = TwoLayerNet(input_size, hidden_size, num_classes)

# Train the network
stats = net.train(X_train, y_train, X_val, y_val,
            num_iters=1000, batch_size=200,
            learning_rate=1e-4, learning_rate_decay=0.95,
            reg=0.25, verbose=True)

# Predict on the validation set
val_acc = (net.predict(X_val) == y_val).mean()
print('Validation accuracy: ', val_acc)
```

**Question**

Now that you have trained a Neural Network classifier, you may find that your testing accuracy is much lower than the training accuracy. Which of the below (possibly multiple) ways can decrease this gap?

1. $\boxed{ \text{Increase the regularization strength.}}$
2. $\boxed{ \text{Train on a larger dataset.}}$
3. Add more hidden units.
4. None of the above.



$\color{blue}{\textit{Answer :}}$ 

The gap between training accuracy and testing accuracy in a Neural Network classifier often indicates overfitting. Overfitting occurs when the model learns to memorize the training data rather than generalize to unseen data. To reduce this gap, we typically need to improve the model's ability to generalize.

Let's analyze each option:

1. **Increase the regularization strength:**
   - **Explanation:** Regularization techniques (such as L2 regularization used in the provided code) penalize large weights in the model. This discourages the model from fitting the noise in the training data excessively and helps in generalizing better to unseen data.
   - **Effect on gap:** Increasing regularization strength typically reduces overfitting by encouraging simpler models (smoother decision boundaries), which can improve testing accuracy relative to training accuracy. Therefore, this option can help decrease the gap between training and testing accuracies.

2. **Train on a larger dataset:**
   - **Explanation:** Training on a larger dataset can provide the model with more diverse examples, which can help it learn more robust features and patterns that generalize better.
   - **Effect on gap:** Training on more data tends to improve the generalization of the model, potentially reducing overfitting. As a result, the testing accuracy may improve relative to the training accuracy, thereby reducing the gap.

3. **Add more hidden units:**
   - **Explanation:** Adding more hidden units increases the capacity of the model, allowing it to learn more complex representations of the data.
   - **Effect on gap:** Adding more hidden units can lead to overfitting if not regularized properly. A more complex model might fit the training data very well (leading to high training accuracy) but may not generalize to unseen data (resulting in lower testing accuracy). Therefore, this option might increase the gap rather than decrease it.

Given these explanations:

- **Option 1 (Increase the regularization strength)** and **Option 2 (Train on a larger dataset)** are likely to decrease the gap between training and testing accuracies by improving generalization.
- **Option 3 (Add more hidden units)** may actually increase the gap because it could exacerbate overfitting unless accompanied by sufficient regularization or more data.

Therefore, the correct answer is:

**1. Increase the regularization strength.**

**2. Train on a larger dataset.**

These options can effectively help in reducing the gap between training and testing accuracies in a Neural Network classifier.

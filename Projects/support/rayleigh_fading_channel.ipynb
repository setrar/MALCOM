###  Rayleigh Fading Channel Models using Julia

```julia
using Plots
```

```julia
# Constants
f_d = 100  # Maximum Doppler shift in Hz
T = 1.0    # Total time in seconds for the simulation
fs = 1000  # Sampling frequency in Hz

# Time vector
t = 0:1/fs:T

# Number of oscillators for Jakes' model
N = 8;
```

```julia
# Generate oscillator phases
phases = rand(N) * 2 * π

# Generate the in-phase and quadrature components
in_phase = sqrt(2/N) * sum(cos.(2π * f_d * cos.(2π * n/N) * t .+ phases[n]) for n = 1:N)
quadrature = sqrt(2/N) * sum(sin.(2π * f_d * cos.(2π * n/N) * t .+ phases[n]) for n = 1:N)

# Rayleigh fading envelope
rayleigh_envelope = sqrt.(in_phase.^2 + quadrature.^2);
```

```julia
# Plotting
plot(t, rayleigh_envelope
    , label="Rayleigh Fading"
    , xlabel="Time (s)", ylabel="Amplitude"
    , title="Rayleigh Fading Channel Model Simulation"
)
```
